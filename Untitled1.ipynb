```python
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_excel("hmwk1data.xls",skiprows=6,index_col="Date")
ax = plt.subplot(111)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_ylabel("Exchange Rate")
data.plot(legend=False,figsize=(12,7.416),title="EUR-USD Exchange Rate",ax=ax)
plt.show()
```

```python
import seaborn as sns
data = pd.read_excel("hmwk1data.xls",skiprows=6)
fig, ax = plt.subplots()
fig.set_size_inches(12, 7.416)
sns.set_style("white")
sns.set_context("notebook", font_scale=1.25, rc={"lines.linewidth": 2.5})
sns.despine()
g = sns.lineplot(x="Date", y="Exchange Rate", data=data, ax=ax).set_title("EUR-USD Exchange Rate")
```

```python
data["percent_change"] = data["Exchange Rate"].pct_change()
ax = data.hist("percent_change",grid=False, figsize=(12,7.416))
ax = ax[0]
for i,x in enumerate(ax):

    # Despine
    x.spines['right'].set_visible(False)
    x.spines['top'].set_visible(False)
```

```python
data['L_percent_change'] = data.percent_change.shift(-1)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(12, 7.416)
sns.despine()
sns.scatterplot(x="L_percent_change", y="percent_change", data=data,ax=ax)
plt.show()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
results = smf.ols('percent_change ~ L_percent_change', data=data).fit()
print(results.summary())
predicted = results.predict()
data.dropna(axis=0, inplace=True, subset =["percent_change"])
data['res'] = data['percent_change'] - predicted
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(12, 7.416)
sns.despine()
sns.scatterplot(data['Date'].tolist(), data['res'])
plt.show()
```
